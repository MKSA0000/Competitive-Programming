```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SimplePerceptron():
    def __init__(self,input_dim):
        self.input_dim = input_dim
        self.w = np.random.normal(size=(input_dim,))
        self.b = 0

    def forward(self,x):
        #x is input data
        #propagate by step function
        y = step(np.matmul(self.w,x) + self.b)
        return y
    def compute_deltas(self,x,t):
        # t is answer data
        y = self.forward(x)
        delta = y - t
        dw = delta * x; db = delta
        return dw, db
def step(x):
    return 1 * (x > 0)
```

```python
d = 2; N = 20
mean = 5
x1 = np.random.randn(N//2,d)
x2 = np.random.randn(N//2,d) + np.array([mean,mean])

t1 = np.zeros(N//2)
t2 = np.ones(N//2)

print(x1.shape)
x = np.concatenate([x1,x2],axis=0)
print(x.shape)
t = np.concatenate([t1,t2])
print([x.shape,t.shape])
plt.scatter(x[:,0], x[:,1],c=t)
plt.show()
```

```python
model = SimplePerceptron(input_dim = d)

def compute_loss(dw,db):
    return all(dw == 0)* (db == 0)
def train_step(x,t):
    dw,db = model.compute_deltas(x,t)
    loss = compute_loss(dw,db)
    model.w = model.w - dw
    model.b = model.b - db
    return loss

while(True):
    classified = True
    for i in range(N):
        loss = train_step(x[i],t[i])
        classified *= loss
    if(classified):
        break
```

```python
print([model.w,model.b])
def yarr(xrr):
    res = -(model.w[0] * xrr + model.b) / model.w[1]
    return res
xr = np.arange(x.min(),x.max(),0.01)
yr = yarr(xr)
```

```python
plt.scatter(xr,yr,marker=1)
plt.scatter(x[:,0], x[:,1],c=t)
plt.show()
```
